# Linear Regression with Gradient Descent

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#import the txt file 
data = pd.read_csv("test.txt", sep=',',names=['population','profit'])
```

```python
#quick (sanity) check to see if it imported properly 
data.head()
```

```python
data_x = data['population']
data_y = data['profit']
```

Below is the algorithm for linear regression with gradient descent.
The link below gives a good explanation of the algorithm and application with different data.
https://towardsdatascience.com/linear-regression-using-gradient-descent-in-10-lines-of-code-642f995339c0

```python
def linearRegressionGD(x,y,iterations,learningrate):
    tot=float(len(y))
    m_current = 0
    b_current = 0
    plt.figure(figsize=(10,10))
    for i in range(iterations):
        y_current = (m_current*x) + b_current
        cost = sum([data**2 for data in y-y_current]) / tot
        #print(cost)
        m_gradient = -(2/tot) * sum(x * (y - y_current))
        b_gradient = -(2/tot) *sum(y-y_current)
        m_current = m_current - (learningrate * m_gradient)
        b_current = b_current - (learningrate * b_gradient)
        #plt.plot(x,y_current, color="blue", linewidth=1)
    plt.plot(x,y_current,color="green",linewidth=1)
    plt.scatter(x,y,color='red')
    plt.show()
```

Now that we have our linear regression algorithm we can see how different learning rates and interations can have an effect on the output.

For our first test, let's make the iteration number high and the learning rate low.

```python
linearRegressionGD(data_x,data_y,10000,0.001)
```

In theory, the more iterations and smaller learning rate should create the best output, but this make take too long to process.

Let's make a new function that shows us each step of the algorithm.

```python
def linearRegressionGDstepByStep(x,y,iterations,learningrate):
    tot=float(len(y))
    m_current = 0
    b_current = 0
    plt.figure(figsize=(10,10))
    for i in range(iterations):
        y_current = (m_current*x) + b_current
        cost = sum([data**2 for data in y-y_current]) / tot
        print(cost)
        m_gradient = -(2/tot) * sum(x * (y - y_current))
        b_gradient = -(2/tot) *sum(y-y_current)
        m_current = m_current - (learningrate * m_gradient)
        b_current = b_current - (learningrate * b_gradient)
        plt.plot(x,y_current, color="blue", linewidth=1)
    plt.plot(x,y_current,color="blue",linewidth=1)
    plt.scatter(x,y,color='red')
    plt.show()
```

Let's see what happens when we set a small learning rate and a low number of iterations.

```python
linearRegressionGDstepByStep(data_x,data_y,5,0.001)
```

As you can see, the lines are approaching a slope similar to the green line in the previous graph. The slope of the lines only change a small amount because our learning rate is small. Next we will increase the numbner of iterations.

```python
linearRegressionGDstepByStep(data_x,data_y,100,0.001)
```

Here, we can see from the output of our cost that we are approaching the minimum cost.
You will find that the higher number of iterations you use, the smaller the cost function will be affected.
You can also "overshoot" the minimum cost if your learning rate is too large.
